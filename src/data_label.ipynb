```python
import json
import time
import pandas as pd
import numpy as np
import csv
pd.options.display.max_colwidth = 200
import random
from IPython.display import clear_output
```

```python
# load tweets data(json)
filename = "../data/HurricaneHarvey.json"
i = 0
with open('output.csv', 'w', newline='',encoding="UTF-8") as csvfile:
    with open(filename, 'r',encoding="UTF-8") as f:
        writer = csv.writer(csvfile, delimiter='=')
        for line in f:
            a = time.time()
            data = json.loads(line)
            writer.writerow([str(data['id']),str(data['text']).strip()])
            i += 1
            if i % 10000 == 0:
                b = time.time()
                print('%i/%i finished (%.8fs per tweet)' % (i,7041866,(b-a)/10000))
```
